The code below performs a prediction using the pretrained imagenet model, creates an adversarial example and also a targeted attack (using FGSM). We do the same using Foolbox, to get an idea how foolbox makes it easy.

```python
from PIL import Image
from torchvision import transforms
import matplotlib.pyplot as plt

# read the image, resize it and convert to Tensor
pig_img = Image.open('pig.jpg')

preprocess = transforms.Compose([
    transforms.Resize(224),
    transforms.ToTensor(),
])

pig_tensor = preprocess(pig_img)[None, :, :, :]
print(pig_tensor.shape)

plt.imshow(pig_tensor[0].numpy().transpose(1, 2, 0))
```

```python
import torch
import torch.nn as nn
from torchvision.models import resnet50
import json

# simple module to normalize an image
class Normalize(nn.Module):
    def __init__(self, mean, std):
        super(Normalize, self).__init__()
        self.mean = torch.Tensor(mean)
        self.std = torch.Tensor(std)
    def forward(self, x):
        # To ensure mean and 'x' are of same type (int, float32, floar64 etc) using type_as method
        return (x - self.mean.type_as(x)[None, :, None, None]) / self.std.type_as(x)[None, :, None, None]

norm = Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])

# Load our model and put it into evaluation mode
model = resnet50(pretrained=True)
model.eval()

# form predictions
pred = model(norm(pig_tensor))

with open("imagenet_class_index.json") as f:
    imagenet_classes = {int(i):x[1] for i,x in json.load(f).items()}
print(imagenet_classes[pred.max(dim=1)[1].item()])
```

```python
# Do the same using foolbox
import foolbox as fb
fb.plot.images(pig_tensor)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#print("using {} ".format(device))

preprocessing = dict(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], axis=-3) # axis = -3 refers to the channels class. channels is generally 3rd from back
bounds = (0, 1)
fmodel = fb.PyTorchModel(model, bounds = (0, 1), preprocessing=preprocessing)

pig = pig_tensor.to(device) # Getting error if we don't use cuda
predictions = fmodel(pig)

print('predicted class is:', imagenet_classes[predictions.max(dim=1)[1].item()])
```

```python
import torch.optim as optim
epsilon = 2. / 255

delta = torch.zeros_like(pig_tensor, requires_grad=True)

opt = optim.SGD([delta], lr=1e-1)

model = model.to(device)
pig_tensor = pig_tensor.to(device) 
for t in range(30):
    # Compute the predictions and loss
    pred = model(norm(pig_tensor + delta))
    loss = -nn.CrossEntropyLoss()(pred, torch.LongTensor([341]).to(device))
    # Display the loss for every 5 epochs
    if t % 5 == 0:
        print(t, '\t', loss.item())
    # Zero the gradients that have been accumilated
    opt.zero_grad()
    # Calculate the gradients
    loss.backward()
    # Take a step using the gradients calculated above
    opt.step()
    # Ensure delta stays within the l-inf bound
    delta.data.clamp_(-epsilon, epsilon)
    
print("True class probability:", nn.Softmax(dim=1)(pred)[0,341].item())
```
